```python
rest_api_key = '7c46340026df0cc06457d86a23463af3'
header = {'Authorization': 'KakaoAK ' + rest_api_key}

for i in hollys_ulsan.index:
    address = hollys_ulsan.loc[i,'address']
    print(address)
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    resp = requests.get(url, headers=header)
    jsondata = resp.json()
    if len(jsondata["documents"])<1:
        continue
    x = jsondata["documents"][0]["address"]["y"]
    hollys_ulsan.loc[i,'위도'] = x
    y = jsondata["documents"][0]["address"]["x"]
    hollys_ulsan.loc[i, '경도'] = y
    print(x, y)
    print()

df
```

```python
hollys_ulsan.dropna(axis=0,inplace=True)
hollys_ulsan
```

```python
map_ulsan = folium.Map(location=[35.542817, 129.323622], zoom_start=12, tiles='OpenStreetMap')
for i in hollys_ulsan.index:
    lat = hollys_ulsan.loc[i, '위도']
    lng = hollys_ulsan.loc[i, '경도']
    contents = folium.Popup('<b>'+df.loc[i, 'store']+'</b>'+'<br>'+df.loc[i, 'address']+'<br>'+str(df.loc[i, 'phone']), max_width=300)
    marker = folium.Marker([lat, lng], popup=contents, icon=folium.Icon(icon='home', color='purple'))
    marker.add_to(map_ulsan)

map_ulsan.save('울산_할리스커피.html')
map_ulsan
```
