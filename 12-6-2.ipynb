```python
map_seoul = folium.Map(location=[37.541, 126.986], zoom_start=12, tiles='OpenStreetMap')
for i in hollys_seoul.index:
    lat = hollys_seoul.loc[i, '위도']
    lng = hollys_seoul.loc[i, '경도']
    service = ""  
    for col in ['dt', '24h', 'parking', 'smoke', 'terrace']:
        if hollys_seoul.loc[i,col]==True:
            if service!="":service+=", "
            tmp = ""
            if col=="dt":
                tmp = "DT 매장 "
            elif col=="24h":
                tmp = "24시간 "
            elif col=="parking":
                tmp = "주차 "
            elif col=="smoke":
                tmp = "흡연시설 "
            elif col=="terrace":
                tmp = "테라스 "
            service+=tmp
    if tmp!="":
            contents = folium.Popup('<b>'+hollys_seoul.loc[i, 'store']+'</b>'+'<br>'+hollys_seoul.loc[i, 'address']+'<br>'+str(hollys_seoul.loc[i, 'phone']+'<br>'+'매장 서비스 : '+service), max_width=300)
    else :contents = folium.Popup('<b>'+hollys_seoul.loc[i, 'store']+'</b>'+'<br>'+hollys_seoul.loc[i, 'address']+'<br>'+str(hollys_seoul.loc[i, 'phone']), max_width=300)
    marker = folium.Marker([lat, lng], popup=contents)
    marker.add_to(map_seoul)

map_seoul.save('서울_할리스커피.html')
map_seoul
```
