# Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Matplotlib   
-    : Object-oriented (OO) style
- pyplot  : pyplot-style

#### Object-Oriented Style

````python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes...
ax.plot(x, x**3, label='cubic') # ... and some more
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.set_title("Simple Plot") # Add a title to the axes. 
ax.legend() # Add a legend.
plt.show()
````

#### pyplot Style
```Python
plot([x], y, [fmt], *, data=None, **kwargs)
plt([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```
- [fmt]
    - ,  ,     
    - fmt='[marker][line][color]'
- **kwargs
    - Line  

````python
plt.plot(x, x, label='linear') # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic') # etc.
plt.plot(x, x**3, label='cubic') 
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
````

````python
plt.plot(x, x, color='green', label='linear')
plt.plot(x, x**2, linestyle='--', label='quadratic')
plt.plot(x, x**3, '^r:', label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()    
plt.show()
````

####

````python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.xlim(0.0, 3.0)
plt.ylim(0, 10)
plt.show()
````

```Python

```

subplot(2,3,4) = subplot(234)

````python
x1 = np.linspace(0, 5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2,y2, '.-')
plt.xlabel('time (s)')  
plt.ylabel('Undamped')

plt.show()
````

####   I
- `subplots()` : Figure Axes  
- `twinx()` :  y  Axes   

- ax0 :   y( )
- ax1 :   y( )  
  
    x ,   y  .

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.bar(bydate.index, bydate.values, label='  ', color='r')
ax1.plot(bydate.index, bydate.values.cumsum(), label='  ', color='g')
plt.xlabel('')
plt.ylabel(' ()')
plt.title('   ')
plt.legend()
plt.show()
````

####   II
- `loc=` :    (**default: 'best'**)

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.set_title('   ')
ax0.bar(bydate.index, bydate.values, label='  ', color='r')
ax0.set_xlabel('')
ax0.set_ylabel(' ()', color='red')
ax0.legend(loc='upper left')

ax1.plot(bydate.index, bydate.values.cumsum(), label='  ', color='g')
ax1.set_ylabel('  ()', color='green')
ax1.legend(loc='upper right')
````

####       I

```Python
plt.pie(x, autopct=None, shadow=False, explode=None, startangle=None, label=None)
```  

- `x` :      
 
- `autopct` :      (`%.2f` :   , `%%` : % )
- `shadow` :    
- `explode` :     (`0`= )(   )
- `startangle` :      ( 0 : 3)   
- `labels` :   

````python
plt.pie(byregion.values, autopct='%.2f%%', shadow=True, explode=(0.1, 0, 0, 0, 0), labels=byregion.index)
plt.title('   ')
plt.legend(loc='upper left')
plt.show()
````
