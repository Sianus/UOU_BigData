# Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Matplotlib를 이용하는 두가지 방법
- 객체 지향 인터페이스 : Object-oriented (OO) style
- pyplot 인터페이스 : pyplot-style

#### Object-Oriented Style

````python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes...
ax.plot(x, x**3, label='cubic') # ... and some more
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.set_title("Simple Plot") # Add a title to the axes. 
ax.legend() # Add a legend.
plt.show()
````

#### pyplot Style
```Python
plot([x], y, [fmt], *, data=None, **kwargs)
plt([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```
- [fmt]
    - 색깔, 마커의 종류, 선의 종류를 문자열 형태로 지정
    - fmt='[marker][line][color]'
- **kwargs
    - Line에 대한 속성

````python
plt.plot(x, x, label='linear') # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic') # etc.
plt.plot(x, x**3, label='cubic') 
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
````

````python
plt.plot(x, x, color='green', label='linear')
plt.plot(x, x**2, linestyle='--', label='quadratic')
plt.plot(x, x**3, '^r:', label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()    
plt.show()
````

````python
x1 = np.linspace(0, 5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2,y2, '.-')
plt.xlabel('time (s)')  
plt.ylabel('Undamped')

plt.show()
````

#### 보조축 사용하기 I
- `subplots()` : Figure와 Axes를 리턴 
- `twinx()` : 보조 y축을 사용하는 Axes 생성  

- ax0 : 첫 번째 y축(일반적으로 왼쪽)
- ax1 : 두 번째 y축(일반적으로 오른쪽)  
  
일반적으로 보조축이라 함은 동일한 x축을 공유하지만, 각각 다른 y축을 가지는 것이다.

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.bar(bydate.index, bydate.values, label='일일 확진자 수', color='r')
ax1.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.title('울산 일자별 확진자 추이')
plt.legend()
plt.show()
````

#### 보조축 사용하기 II
- `loc=` : 범례 위치 지정 (**default: 'best'**)

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.set_title('울산 일자별 확진자 추이')
ax0.bar(bydate.index, bydate.values, label='일일 확진자 수', color='r')
ax0.set_xlabel('일자')
ax0.set_ylabel('확진자 수(명)', color='red')
ax0.legend(loc='upper left')

ax1.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
ax1.set_ylabel('누적 확진자 수(명)', color='green')
ax1.legend(loc='upper right')
````

#### 지역별 확진자 수 파이형 차트 그리기 I

```Python
plt.pie(x, autopct=None, shadow=False, explode=None, startangle=None, label=None)
```  

- `x` : 파이 차트를 구성할 데이터  
 
- `autopct` : 파이 조각의 전체 대비 백분율 (`%.2f` : 소수점 둘째자리까지 실수, `%%` : %기호 표시)
- `shadow` : 그림자 효과 유무 
- `explode` : 파이 조각의 돌출 정도 (`0`= 돌출안됨)(파이조각의 개수만큼 개별 설정)
- `startangle` : 파이 조각의 시작 위치의 각도 (기본값 0 : 3시방향) 반시계방향으로 각도만큼 이동
- `labels` : 파이 조각의 라벨

````python
plt.pie(byregion.values, autopct='%.2f%%', shadow=True, explode=(0.1, 0, 0, 0, 0), labels=byregion.index)
plt.title('울산 지역별 확진자 비율')
plt.legend(loc='upper left')
plt.show()
````

#### 관계형 데이터와 범주형 데이터 구분 요령
- **자료의 성격** : 데이터가 연속적인 숫자로 이루어져 있다면 관계형 데이터, 명확한 범주로 나누어진다면 범주형 데이터 

- **분석 목적** : 두 변수 사이의 관계를 보고 싶다면 관계형 데이터, 그룹 간 비교나 빈도수를 보고 싶다면 범주형 데이터

### Figure-level vs. Axes-level functions

  
- Figure-level Functions
    - `replot`, `displot`, `catplot` 과 같은 함수들이 있음  

    - 전체 그림(figure)과 그 안에 포함된 모든 축(axes)을 관리한다. 
    - 복잡한 레이아웃을 가진 여러 그래프를 한 번에 그리는 데 적합하다.  
    - 자체적으로 `Figure` 와 `Axes`를 생성하고 관리한다.

- Axes-level Functions
    - `scatterplot`, `lineplot`, `histplot` 등이 있다.  

    - 특정 축(axes)에 그래프를 그리고, 하나의 축에 대해 더 세밀한 제어를 할 수 있다.
    - 이미 존재하는 `Axes`에 그래프를 그리는 데 사용된다. 
    - 해당 함수들을 사용할 때는 사용자가 `plt.subplots()` 등을 통해 `Axes`를 미리 생성하고 관리해야 한다.

#### Relational plots
- 이 함수는 관계형 데이터를 시각화하기 위한 것이다.  

- 데이터 포인트 간의 관계를 나타내는 데 사용된다. 
- `scatterplot` : 두 변수 간의 관계를 점으로 나타낸다. 
- `lineplot` : 두 변수 간의 관게를 선으로 연결하여 나타낸다. 

```Python
replot(kind='scatter') :default = scatterplot()
```

### relplot

### catplot()
- default : `strip`

````python
sns.catplot(x='day', y='total_bill', data=tips, palette='coolwarm')
````

#### kind=box


- **중앙값 (Median)**  

    - 데이터의 중간 값을 나타낸다. 데이터를 작은 순서대로 나열했을 때, 가운데 위치하는 값이다. 
    - 박스 플롯 중앙에 있는 가로선으로 표시된다. 

- **사분위수**
    - 데이터를 네 등분한 값들이다. 
    - Q1(하위 사분위수) : 데이터의 하위 25%에 해당하는 값이다. 박스 플롯의 아래쪽 끝이다. 
    - Q3(상위 사분위수) : 데이터의 상위 25%에 해당하는 값이다. 박스 플롯의 위쪽 끝이다. 

- **사분위수 범위 (IQR)**
    - Q3와 Q1의 차이를 나타내며(Q3-Q1), 데이터의 중간 50%가 얼마나 퍼져 있는지를 보여준다. 

- **최소값 ~ 최대값 (Whiskers)**
    - 박스의 상단과 하단에서 뻗어 나온 선들로, 일반적으로 `Q1-1.5*IQR` 과 `Q3_1.5*IQR` 사이의 값들을 나타낸다. 
    - 이 범위를 벗어나는 데이터는 일반적으로 이상치로 간주된다. 

- **이상치** 
    - Whiskers 범위를 벗어나는 점들로, 개별적으로 표시되며 일반적인 데이터 범위에서 벗어난 값을 나타낸다. 

````python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', palette='coolwarm')
````

````python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='smoker')
````

#### kind=violin

- `violin`은 데이터의 분포와 확률 밀도를 시각화하는 그래프다.  

- **중앙의 흰색 선(두꺼운 선)**
    - 이 선은 박스 플롯의 중앙값(Median)을 나타낸다.  

    - 데이터의 중간 지점을 보여준다. 
- **중앙의 검은색 두꺼운 막대**
    - 박스 플롯의 사분위 범위(IQR)을 나타낸다. 
    
    - 일반적으로 데이터의 중간 50%에 해당하는 Q1부터 Q3까지의 범위를 나타낸다. (25% ~ 75%)
- **중앙의 검은색 얇은 선**
    - 일반적으로 Q1과 Q3 에서 1.5 * IQR 범위 밖의 데이터를 이상치로 간주하기 전까지의 범위를 나타낸다. 

````python
sns.catplot(x='total_bill', y='day', data=tips, kind='violin', hue='sex')
````

````python
sns.catplot(x='total_bill', y='day', data=tips, kind='violin', hue='sex', split=True)
````

### pairplot()
- Dataset의 각 열에 상관 관계를 보여주는 multi-plot

- **대각선 상의 그래프(1행 1열, 2행 2열, 3행 3열)**  

    - 각 변수의 분포를 나타내는 히스토그램  

    - 히스토그램은 해당 변수의 값의 빈도를 표현한다. 
    - 이를 통해 우리는 각 변수의 분포가 어떤 모습인지, 데이터가 어떻게 퍼져 있는지를 볼 수 있다. 

- **대각선을 기준으로 한 아래쪽 산점도(2행 1열, 3행 1열, 3행 2열)**

    - 두 변수 간의 관계를 나타내는 산점도다. 

- **대각선을 기준으로 한 위쪽 산점도(1행 2열, 1행 3열, 2행 3열)** 

    - 두 변수 간의 관계를 나타내는 산점도지만, 반대 방향이다. 

````python
sns.pairplot(tips)
````

- **1행 1열**

    - `total_bill` 변수의 히스토그램을 보여준다. 히스토그램은 해당 변수의 데이터 분포와 빈도를 나타낸다. 
    
    - 이를 통해 `total_bill` 의 데이터가 어떻게 분포되어 있는지, 가장 자주 나타나는 청구 금액 범위가 어디인지, 분포의 모양(대칭적인지, 왜곡되어 있는지)을 볼 수 있다. 

- **1행 2열**

    - `total_bill`과 `tip` 사이의 관계를 나타내는 산점도다. 
    
    - 가로축은 `total_bill`, 세로축은 `tip`을 나타낸다. 
    - 산점도를 통해 두 변수 간의 상관관계를 볼 수 있으며, `total_bill` 이 증가함에 따라 `tip` 도 증가하는 경향이 있음을 알 수 있다. 


- **2행 2열**

    - `tip`의 히스토그램을 보여준다. 
    
    - 즉, `tip`이 얼마나 자주 어떤 금액에 주어지는 지를 시각화 한다. 

````python
sns.pairplot(tips, hue='sex')
````
