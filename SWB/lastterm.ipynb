# Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Matplotlib를 이용하는 두가지 방법
- 객체 지향 인터페이스 : Object-oriented (OO) style
- pyplot 인터페이스 : pyplot-style

#### Object-Oriented Style

````python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes...
ax.plot(x, x**3, label='cubic') # ... and some more
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.set_title("Simple Plot") # Add a title to the axes. 
ax.legend() # Add a legend.
plt.show()
````

#### pyplot Style
```Python
plot([x], y, [fmt], *, data=None, **kwargs)
plt([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
```
- [fmt]
    - 색깔, 마커의 종류, 선의 종류를 문자열 형태로 지정
    - fmt='[marker][line][color]'
- **kwargs
    - Line에 대한 속성

````python
plt.plot(x, x, label='linear') # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic') # etc.
plt.plot(x, x**3, label='cubic') 
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
````

````python
plt.plot(x, x, color='green', label='linear')
plt.plot(x, x**2, linestyle='--', label='quadratic')
plt.plot(x, x**3, '^r:', label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()    
plt.show()
````

#### 축 범위 설정
- 기본은 자동 설정

- xlim(left, right) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  ylim(bottom, top)

````python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.xlim(0.0, 3.0)
plt.ylim(0, 10)
plt.show()
````

### matplotlib.pyplot.subplot : 여러 차트 그리기

```Python
subplot(nrows, ncols, index, **kwargs)
subplot(pos, **kwargs)
subplot(**kwargs)
subplot(ax)
```
- nrows : 행의 수 
- ncols : 열의 수 
- index : 위치(몇번째)

subplot(2,3,4) = subplot(234)

````python
x1 = np.linspace(0, 5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2,y2, '.-')
plt.xlabel('time (s)')  
plt.ylabel('Undamped')

plt.show()
````

#### 일자별 확진자 수 구하기

- `value_counts()` : 유일한 값 별로 개수 구하기  

- `sort_index()` : 인덱스를 기준으로 정렬
    - `ascending=True` : 오름차순(**default**)  
    
    - `ascending=False` : 내림차순
- `sort_values()` : 값을 기준으로 정렬

````python
bydate = df['확진일'].value_counts().sort_index()
bydate
````

````python
bydate.index
````

#### 일자별 확진자 수 꺽은선형 차트 그리기

````python
plt.plot(bydate.index, bydate.values, label='일일 확진자 수', color='r')
plt.title('울산 일자별 확진자 추이')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.legend()
plt.show()
````

#### Figure 크기 조절
- `figsize=` : 생략시 기본 크기(6.4, 4.8)

````python
plt.figure(figsize=(15,5))
plt.plot(bydate.index, bydate.values, label='일일 확진자 수', color='r')
plt.title('울산 일자별 확진자 추이')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.legend()
plt.show()
````

#### 일자별 확진자 수 세로막대형 차트 그리기

````python
plt.figure(figsize=(15,5))
plt.bar(bydate.index, bydate.values, label='누적 확진자 수', color='r')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.title('울산 일자별 확진자 추이')
plt.legend()
plt.show()
````

#### 누적 확진자 수 꺽은선형 차트 그리기
- `cumsum()` : 값의 누적 합을 구하는 함수

````python
plt.figure(figsize=(15,5))
plt.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.legend()
plt.title('울산 일자별 누적 확진자 수 추이')
plt.show()
````

#### 일자별 확진자 수와 누적 확진자 수 동시에 그리기

````python
plt.figure(figsize=(15,5))
plt.bar(bydate.index, bydate.values, label='일일 확진자 수', color='r')
plt.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.title('울산 일자별 확진자 추이')
plt.legend()
plt.show()
````

#### 보조축 사용하기 I
- `subplots()` : Figure와 Axes를 리턴 
- `twinx()` : 보조 y축을 사용하는 Axes 생성  

- ax0 : 첫 번째 y축(일반적으로 왼쪽)
- ax1 : 두 번째 y축(일반적으로 오른쪽)  
  
일반적으로 보조축이라 함은 동일한 x축을 공유하지만, 각각 다른 y축을 가지는 것이다.

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.bar(bydate.index, bydate.values, label='일일 확진자 수', color='r')
ax1.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
plt.xlabel('일자')
plt.ylabel('확진자 수(명)')
plt.title('울산 일자별 확진자 추이')
plt.legend()
plt.show()
````

#### 보조축 사용하기 II
- `loc=` : 범례 위치 지정 (**default: 'best'**)

````python
fig, ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()
ax0.set_title('울산 일자별 확진자 추이')
ax0.bar(bydate.index, bydate.values, label='일일 확진자 수', color='r')
ax0.set_xlabel('일자')
ax0.set_ylabel('확진자 수(명)', color='red')
ax0.legend(loc='upper left')

ax1.plot(bydate.index, bydate.values.cumsum(), label='누적 확진자 수', color='g')
ax1.set_ylabel('누적 확진자 수(명)', color='green')
ax1.legend(loc='upper right')
````

#### 지역별 확진자 수 파이형 차트 그리기 I

```Python
plt.pie(x, autopct=None, shadow=False, explode=None, startangle=None, label=None)
```  

- `x` : 파이 차트를 구성할 데이터  
 
- `autopct` : 파이 조각의 전체 대비 백분율 (`%.2f` : 소수점 둘째자리까지 실수, `%%` : %기호 표시)
- `shadow` : 그림자 효과 유무 
- `explode` : 파이 조각의 돌출 정도 (`0`= 돌출안됨)(파이조각의 개수만큼 개별 설정)
- `startangle` : 파이 조각의 시작 위치의 각도 (기본값 0 : 3시방향) 반시계방향으로 각도만큼 이동
- `labels` : 파이 조각의 라벨

````python
plt.pie(byregion.values, autopct='%.2f%%', shadow=True, explode=(0.1, 0, 0, 0, 0), labels=byregion.index)
plt.title('울산 지역별 확진자 비율')
plt.legend(loc='upper left')
plt.show()
````
