```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label="linear")
ax.plot(x, x*x, label="quadratic")
ax.plot(x, x**3, label="cubic")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
ax.legend()
plt.show()
```

```python
x = np.linspace(0,2,20)

plt.plot(x,x, label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.title("Simple Plot")
plt.show()
```

```python
x = np.linspace(0,2,20)

plt.plot(x,x, color="green", label="linear")
plt.plot(x,x**2, linestyle = "--", label="quadratic")
plt.plot(x,x**3, marker="^", label="cubic")
plt.plot(x,x**4, "k--^", label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.xlim([0,1])
plt.ylim([0,1])
plt.legend()
plt.title("Simple Plot")
plt.show()
```

```python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,2,1)
plt.plot(x1, y1, "o-")
plt.title("A tale of 2 subplots")
plt.ylabel("Damped oscillation")

plt.subplot(2,2,4)
plt.plot(x2, y2, "o-")
plt.xlabel("time (s)")
plt.ylabel("Undamped")
```

```python
x = np.linspace(0,2,20)

plt.subplot(2,3,1)
plt.plot(x,x, label="linear")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("1 Plot")
plt.subplot(2,3,3)
plt.plot(x,x**2,label="quadratic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("2 Plot")
plt.subplot(2,3,5)
plt.plot(x,x**3,label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("3 Plot")
plt.show()
```

```python
import urllib3.exceptions
for i in df.index:
    address = df.loc[i, "주소"]
    url = "https://dapi.kakao.com/v2/local/search/address.json?query=" + address

    rest_api_key = "7ba327055e33d9ffbda555e6b24e8e2b"

    header = {"Authorization" : "KakaoAK " + rest_api_key}
    try:
        resp = requests.get(url, headers=header)
    except ConnectionResetError as e:
        print("Error on " + address + e)
    except requests.exceptions.ConnectionError as e:
        print("Error on " + address + str(e))
    except urllib3.exceptions.ProtocolError:
        print("Error on " + address + e)

    if resp.status_code == 200:
        jsondata = resp.json()

        if jsondata["meta"]["total_count"] == 0: continue

        df.loc[i, "경도"] = jsondata["documents"][0]["address"]["x"]
        df.loc[i, "위도"] = jsondata["documents"][0]["address"]["y"]
    else:
        df.loc[i, "경도"] = "ERROR" + str(resp.status_code)
        df.loc[i, "위도"] = "ERROR" + str(resp.status_code)
df
```

```python
titles = soup.select("div > div > article > a > div.card-photo > img")                      # 물건 이름 가져오기
regions = soup.select("div > div > article > a > div.article-info > p.article-region-name") # 지역 이름 가져오기
prices = soup.select("div > div > article > a > div.article-info > p.article-price")        # 가격 가져오기

titlelist = []
regionlist = []
pricelist = []

for title, region, price in zip(titles, regions, prices):
	dtitle = title.get("alt")
	dreg = region.text.replace("\n", "").strip()
	dprice = price.text.replace("\n", "").strip()
	titlelist.append(dtitle)
	regionlist.append(dreg)
	pricelist.append(dprice)

dicResult = {"물품명" : titlelist, "지역" : regionlist, "가격" : pricelist}
```

```python
dfResult = pd.DataFrame(dicResult)
dfResult
```

```python
dfResult["가격"] = dfResult["가격"].str.replace(",","").str.replace("만", "0000").str.replace("원", "").astype(int)
dfResult
```

```python
dfResult["구"] = dfResult["지역"].apply(lambda x:x.split()[1])
dfResult["동"] = dfResult["지역"].apply(lambda x:x.split()[2])
dfResult
```

```python
result = []

for i in range(1,51):
    url = "https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store="%i
    resp = requests.get(url)

    soup = BeautifulSoup(resp.text, "html.parser")

    trs = soup.find_all("tr")
    for tr in trs[1:]:
        td = tr.find_all("td")
        store_reg = td[0].string
        store_name = td[1].string
        store_addr = td[3].string
        store_phone = td[5].string
        store_dt = True if len(tr.find_all("img", attrs={"alt":"DT 매장"}))!=0 else False
        store_terrace = True if len(tr.find_all("img", attrs={"alt":"테라스"}))!=0 else False
        store_24 = True if len(tr.find_all("img", attrs={"alt":"24시간"}))!=0 else False
        store_smoke = True if len(tr.find_all("img", attrs={"alt":"흡연시설"}))!=0 else False
        store_park = True if len(tr.find_all("img", attrs={"alt":"주차"}))!=0 else False
        result.append([store_name, store_reg, store_addr, store_phone, store_dt, store_terrace, store_24, store_smoke, store_park])

dfResult = pd.DataFrame(data=result, columns=["매장명", "지역", "주소", "전화번호", "DT 매장", "테라스", "24시간", "흡연시설", "주차"])
dfResult
```
