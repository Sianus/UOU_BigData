```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/ASOS_month(종관기상관측)).csv",encoding='cp949')
df
```

```python
df.isna().sum()
```

```python
df['평균기온(°C)'] = df['평균기온(°C)'].interpolate(method='linear')
```

```python
df['연'] = df['일시'].apply(lambda x:x.split('-')[0])
df
```

```python
df1 = df[['평균기온(°C)','연']].groupby('연').mean()
df1
```

```python
df.loc[11]
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data = df1)
```
